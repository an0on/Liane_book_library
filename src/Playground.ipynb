```python
query_active_loans = """
    SELECT 
        borrowers.FirstName, 
        borrowers.LastName, 
        books.Title, 
        loans.BorrowDate
    FROM loans
    JOIN borrowers ON loans.BorrowerID = borrowers.BorrowerID
    JOIN books ON loans.BookID = books.BookID;
"""

loans_df = pd.read_sql(query_active_loans, con=connection_string)
loans_df
```
